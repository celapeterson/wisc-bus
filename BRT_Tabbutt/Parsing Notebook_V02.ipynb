## Make a JSON file for looking up latitude and longitude of stop IDs

```python
# Make a dictionary of stop Ids and lat and long for them:

stopIDLookUp = {}

for i in stopsData['stop_id']:
    idx = stopsData.loc[stopsData['stop_id'] == i].index[0]
    stopIDLookUp[i] = [stopsData['stop_lat'][idx], stopsData['stop_lon'][idx]]

with open('stopIDLookUp.json', 'w') as fp:
    json.dump(stopIDLookUp, fp)
```

## Make a JSON file for looking up latitude and longitude of stop IDs (BRT)

```python
BRTStops = pd.read_csv('Bus_Rapid_Transit_Stops.csv', sep=",")
BRTStops.head(3)
```

```python
stopIDLookUpBRT = {}

for i in BRTStops['OBJECTID']:
    idx = BRTStops.loc[BRTStops['OBJECTID'] == i].index[0]
    stopIDLookUpBRT[i] = [BRTStops['stop_lat'][idx], BRTStops['stop_lon'][idx]]
    
with open('stopIDLookUpBRT.json', 'w') as fp:
    json.dump(stopIDLookUpBRT, fp)
```

## Want to make a couple JSONs to convert route name to tripID, and list every possible stop for each route

```python
routeNameToTripID = {}

for i in tripsData['trip_id']:
    idx = tripsData.loc[tripsData['trip_id'] == i].index[0]
    routeNameToTripID[i] = int(tripsData['route_short_name'][idx])
    
with open('routeNameToTripID.json', 'w') as fp:
    json.dump(routeNameToTripID, fp)
```

```python
routesStopsTryAgain = {}
routeCounter = []

for i in stopTimesData.iterrows():
    if routeNameToTripID[i[1]['trip_id']] in routesStopsTryAgain:
        if i[1]['stop_id'] in routesStopsTryAgain[routeNameToTripID[i[1]['trip_id']]]:
            continue
        else:
            routesStopsTryAgain[routeNameToTripID[i[1]['trip_id']]].append(int(i[1]['stop_id']))
    else: 
        routesStopsTryAgain[routeNameToTripID[i[1]['trip_id']]] = []
        routesStopsTryAgain[routeNameToTripID[i[1]['trip_id']]].append(int(i[1]['stop_id']))
        
with open('routesStopsTryAgain.json', 'w') as fp:
    json.dump(routesStopsTryAgain, fp)
```
